#### 2: Write a user-defined function for origin  


```python
#use apply function
def origin_name(number):
    og = ''
    
    if (number==1):
        og = 'House'
    elif (number==2):
        og = 'Campus'
    elif (number==3):
        og = 'Heaven'
        
    return og
    
```

```python
#view first 30 data points 
auto_df['origin'] = auto_df['origin'].apply(origin_name)
auto_df['origin'].head()
```

#### 3: Draw the pair plot using sns for mpg, weight,orign and with hue origin, set the size to 4


```python
plt.figure(figsize=(4,4))
selection = ['mpg', 'weight', 'origin']
auto_df_limited = auto_df[selection]
sns.pairplot(auto_df_limited, hue="origin")
```
