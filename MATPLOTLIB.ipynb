# MATPLOTLIB

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
randomNumber = np.random.rand(10)
print(randomNumber)
```

```python
style.use('ggplot')
plt.plot(randomNumber, 'g', label='line 1', linewidth=2)
plt.xlabel('Range')
plt.ylabel('Numbers')
plt.title('First Plot')
plt.legend()
plt.show
```

```python
time = range(0,24)
people = [np.random.randint(0, 100) for x in range(24)]
```

```python
plt.plot(time, people, 'r', label='web traffic', linewidth=2)
plt.xlabel('Hrs')
plt.ylabel('People')
plt.title('Web traffic')
```

```python
plt.plot(time, people, alpha=.4)
plt.annotate('Max', ha='center', va='bottom',
             xytext=(5,60), xy=(time[people.index(max(people))],max(people)), 
             arrowprops={'facecolor':'green'})
```

```python
# data
time_hrs = range(0,24)
temp = [np.random.randint(0, 110) for x in range(24)]
wind = [np.random.randint(0, 20) for x in range(24)]
humidity = [np.random.randint(60, 99) for x in range(24)]
precipitation = [np.random.randint(0, 100) for x in range(24)]
```

```python
plt.figure(figsize=(8,4))
plt.subplots_adjust(hspace=.25)
plt.subplot(1,2,1)
plt.title('Temp')
plt.plot(time_hrs, temp, color='b', linestyle='-', linewidth=2)
plt.subplot(1,2,2)
plt.title('Wind')
plt.plot(time_hrs, wind, color='g', linestyle='-', linewidth=2)
```

```python
plt.figure(figsize=(8,4))
plt.subplots_adjust(hspace=.25)
plt.subplot(1,2,1)
plt.title('Humidity')
plt.plot(time_hrs, humidity, color='b', linestyle='-', linewidth=2)
plt.title('Precipitation')
plt.plot(time_hrs, precipitation, color='r', linestyle='-', linewidth=2)
```

```python
plt.figure(figsize=(8,4))
plt.subplots_adjust(hspace=.5)
plt.subplot(2,2,1)
plt.title('Temp')
plt.plot(time_hrs, temp, color='b', linestyle='-', linewidth=2)
plt.subplot(2,2,2)
plt.title('Wind')
plt.plot(time_hrs, wind, color='g', linestyle='-', linewidth=2)
plt.subplot(2,2,3)
plt.title('Humidity')
plt.plot(time_hrs, humidity, color='y', linestyle='-', linewidth=2)
plt.subplot(2,2,4)
plt.title('Precipitation')
plt.plot(time_hrs, precipitation, color='r', linestyle='-', linewidth=2)
```
